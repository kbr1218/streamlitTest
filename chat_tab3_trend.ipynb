## 01. 데이터 가져오기

```python
df = pd.read_csv('data/JEJU_MCT_DATA_v2.csv', encoding='cp949')
df.head(3)
```

```python
menu_type = df['MCT_TYPE'].value_counts()
menu_type
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=menu_type.index, y=menu_type.values, alpha = 0.8)
plt.xticks(rotation = -45)
sns.set(font_scale=0.6) 
plt.show()
```

```python
plt.close('all')
```
